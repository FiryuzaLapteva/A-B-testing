### Задача 1
1)Вы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты по метрике конверсий в покупку. Выберите метод оценки и оцените есть ли стат.значимые различия между конверсиями в двух группах при alpha = 5%. Дайте краткие рекомендации команде .Результаты: 1)Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550 . 2) Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента: n1 = 164 и n2 = 228 3) Конверсии : conv1 = 1.05% conv2 = 1.47% . 

Для сравнения долей используют z критерий, который находят в случае сравнения  долей по формуле:

![Формула Z - критерия](\images\picture.jpg)

где n_1 и n_2 – объемы выборок, а m_1 и m_2 – число интересующих объектов в выборке.

Величина Z – это величина, которая следует нормальному стандартному распределению, но выше вычисленная величина z немного отклоняется от нормального распределения, поэтому  при сравнении долей нам нужно воспользоваться поправкой Йейтса на непрерывность. Данная поправка сокращает отклонения от нормальности и z критерий будем в итоге находить по формуле:

![Формула Z - критерия с поправкой Йейтса](\images\picture2.jpg)



```python
from math import sqrt, fabs
# Дано
x1 = 164 # число покупок в первой группе
x2 =228 # число покупок во второй группе
n1=15550 # выборка 1
n2=15550 # выборка 2
p1 = x1/n1 # доля в первой выборке
p2 = x2/n2 # доля во второй выборке
p = (x1+x2)/(n1 + n2) # общая доля интересующих объектов

results2 =(p1-p2)/sqrt(p*(1-p)*(1/n1 + 1/n2))
print(f'Z - критерий {results2}')

#results = fabs(p1-p2) - 0.5*(1/n1 + 1/n2)/ sqrt(p*(1-p)* (1/n1 + 1/n2))
#print(f'Z - критерий с поправкой на Йeтса = {results}')
```

Для двустороннего теста табличный критерий  z_(( α)/2)=±1.96 для α=5% 

Все что находится за этими табличными значениями, попадает в область принятия гипотезы H_1, в нашем случае Z - критерий с поправкой на Йетси = -0.0467, значение попадает в область [-1,96; +1,96], область принятия гипотезы H_0.
Соответсвенно статистически значимых различий в конверсии двух групп нет.
